```python
N = 580
Q = 12
Users1 = []
Scenarios1 = []
Features_Usa_Sec = []
for i in range(1,N+1):
    for j in range(1,Q+1):
        for k in range(0,2):
            Users1.append(i)
            Scenarios1.append(j)
            Features_Usa_Sec.append(k)
print(Scenarios1)
```
