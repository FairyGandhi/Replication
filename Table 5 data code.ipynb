```python
df['HomeExpDum'] = df['HomeExpDum'].fillna(5)
df['MedRecDum'] = df['MedRecDum'].fillna(5)
df['PwdDum'] = df['PwdDum'].fillna(5)
```

```python
df
```

```python
# HomeExpDum = 1, for n = 4,640
max_iterations = 1
n = 4640
i = 1 #counter
while True:
    df_n = df.sample(n)
    df_n.loc[df_n['HomeExpDum'] == 5, 'HomeExpDum'] = 1
    i+=1
    if i>max_iterations:
        break
df_n
```

```python
df.loc[df_n.index, :] = df_n[:]  
df
```

```python
# MedRecDum = 1, for n = 4,640
max_iterations = 1
n = 4640
i = 1 #counter
while True:
    df_n = df[(df.HomeExpDum == 5) & (df.MedRecDum == 5)].sample(n)
    df_n.loc[df_n['HomeExpDum'] == 5, 'MedRecDum'] = 1
    i+=1
    if i>max_iterations:
        break
df_n
```

```python
df.loc[df_n.index, :] = df_n[:]  
df
```
