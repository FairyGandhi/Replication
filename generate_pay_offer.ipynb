Regenerating the data structure from [paper]

```python
import pandas as pd
import random
```

```python
df = pd.read_csv('replication_data_df_new.csv', index_col=0)

# We're going to regenerate this column...
df = df.drop('Pay_Offer', axis=1)
df
```

```python
# We want Pay_Offer to range between [0.24, 2.99, step=0.25]
i = 24
possible_pay_offers = range(i, i*(12 + 1), 25)
possible_pay_offers

def random_pay_offer():
    return random.choice(possible_pay_offers) / 100
```

```python
grouped_pay_offer = df.groupby(['User','Scenario']).apply(lambda x: random_pay_offer())
grouped_pay_offer.name = 'Pay_Offer'
grouped_pay_offer
```

```python
df.new = df.join(other=grouped_pay_offer, on=['User','Scenario'])
```

```python
import csv
df.new.to_csv('replication_data_df_new_updated.csv')
```
